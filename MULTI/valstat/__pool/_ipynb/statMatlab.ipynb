# Matlab - néhány egyszerű dolog

* normális vektorok generálása `randn`
* reprodukálhatóság: 'seed'
* átlag: `mean(x)`$=\overline{x}=\frac{x_1+\ldots+x_n}{n}$

```octave
randn('seed',42); % reprodukálhatóság
n=1000;
x=randn(n,1); % ; a végére és nem "írja" ki (quiet)
y=randn(n,1);
mean(x)
mean(y)
```

1. maximális, minimális érték: `max` ill. `min`
1. középső érték, medián: `median`
1. korrigált ($n-1$ a nevezőben) tapasztalati szórás: `std(x)`$=\frac{(x_1-\overline{x})^2+\ldots+(x_n-\overline{x})^2}{n-1}$
1. rang, terjedelem: `range`=`max`-`min`

```octave
max(x)
min(y)
median(x)
std(x)
range(x)
range(y)==max(y)-min(y) % 1 == true, 0 == false
```

* rendezés: `sort`
* kvantilisek: a vektor elemeit közelítőleg $q\%:(100-q\%)$ arányban felosztó szám: `prctile(x,q)` 

```octave
median(y)==prctile(y,50)
prctile(x,[1,10,50,90,99])
```

* histogram: `hist`
* elméleti sűrűségfv: `normpdf` (probability density fun.)


```octave
hist(x)
```

* közelítés: $F(x+\Delta x)-F(x)\approx f(x)\Delta x \approx f(x+\frac{\Delta x}{2})\Delta x$

```octave
nbin=10
hist(x,nbin,1) %6 részre osztja a tartományt, 1-re normálja az összterületet
% lásd help hist
hold on
h=range(x)/nbin;
dom=min(x)+h/2:h/2:max(x)-h/2;
plot(dom,h*normpdf(dom,mean(x),std(x)))
```
